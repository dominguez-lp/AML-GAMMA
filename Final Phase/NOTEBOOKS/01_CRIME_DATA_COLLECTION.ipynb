```python
def build_agency_df(data, ori):

    if 'actuals' not in data or len(data['actuals']) == 0:
        return None

    agency_label = next(iter(data['actuals']))

    rates = data['rates'].get(agency_label, {})
    actuals = data['actuals'].get(agency_label, {})

    agency_name = agency_label.replace(" Arrests","")

    population = data['populations']['population'].get(agency_name, {})
    participated_pop = data['populations']['participated_population'].get(agency_name, {})

    df = pd.DataFrame({
        'month': rates.keys(),
        'actual_arrests': actuals.values(),
        'rate_per_100k': rates.values(),
        'population': population.values(),
        'participated_population': participated_pop.values()
    })

    df['month'] = pd.to_datetime(df['month'], format='%m-%Y')
    df['ori'] = ori

    return df.sort_values('month').reset_index(drop=True)
```
