```python
import plotly.express as px

fig = px.scatter_mapbox(
    df_NM,
    lat="latitude",
    lon="longitude",
    size="Arrests_2024",          # bubble size
    color="Arrests_2024",         # optional color scaling
    hover_name="agency_name",
    hover_data=["Arrests_2024", "Population_2024"],
    size_max=40,
    zoom=6
)

fig.update_layout(
    mapbox_style="carto-positron",  # clean background
    margin={"r":0,"t":0,"l":0,"b":0}
)

plt.savefig("../FIGURES/Figure_1.png")

fig.show()
```

```python
with open('../DATA/RAW_CRIME_DATA/response_1771464984839.json', 'r') as f:
    data = json.load(f)

actuals = data["actuals"]["United States Arrests"]
population = data["populations"]["population"]["United States"]
rates = data["rates"]["United States Arrests"]

df = pd.DataFrame({
    "Month": actuals.keys(),
    "Arrests": actuals.values()
})

df["Population"] = df["Month"].map(population)
df["Arrest_Rate_per_100k"] = df["Month"].map(rates)

df
```
